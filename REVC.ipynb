```python
with open("C:\\Users\\ABCLAB\\Rosalind\\data\\rosalind_revc.txt",'r') as f:
    line = f.readline()
    
print(line)
```

```python
s_dna = []
```

```python
for i in reversed(range(len(line))):
    if line[i] == 'A':
        s_dna.append('T') 
    elif line[i] == 'C':
        s_dna.append('G') 
    elif line[i] == 'T':
        s_dna.append('A') 
    elif line[i] == 'G':
        s_dna.append('C') 
```

```python
s_dna = ''.join(s_dna)
```
