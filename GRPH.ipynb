```python
with open("C:\\Users\\ABCLAB\\Rosalind\\data\\rosalind_grph.txt",'r') as f:
    line = f.readlines()

#rna_str = [line[i:i+3] for i in range(0,len(line),3)]
#print(*line)
```

```python
id_idx = []

for i in range(len(line)):
    if line[i].find('>') != -1:
        id_idx.append(i)
```

```python
dna_str = []

for i in range(len(id_idx)-1):
    dna_str.append(''.join(line[id_idx[i]+1:id_idx[i+1]]))

dna_str.append(''.join(line[id_idx[len(id_idx)-1]+1:len(line)]))

for i in range(len(dna_str)):
    dna_str[i] = dna_str[i].replace('\n',"")
#print(dna_str)
```

```python
dna_id = [line[id_idx[i]].strip('>').strip() for i in range(len(id_idx))]
#print(dna_id)
```

```python
adj_list = []
for i in range(len(dna_str)):
    for j in range(len(dna_str)):
        if i == j:
            continue
        elif dna_str[i][-1] == dna_str[j][0]:
            adj_list.append([dna_id[i], dna_id[j]])
```
