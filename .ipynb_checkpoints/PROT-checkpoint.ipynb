```python
with open("C:\\Users\\ABCLAB\\Rosalind\\data\\rna_codon_table.txt",'r') as f:
    codon_table = f.readlines()
    
print(codon_table)
```

```python
for i in range(len(codon_table)):
    codon_table[i] = codon_table[i].strip()
```

```python
codon_table_v2 = []

for i in range(len(codon_table)):
    codon_table_v2 += codon_table[i].split(' '*6)
print(codon_table_v2)
    
print('-'*10)
for i in range(len(codon_table_v2)):
    codon_table_v2 += codon_table_v2[i].split(' '*3)
print(codon_table_v2)
```

```python
codon_table_dic = {}

for i in range(len(codon_table_v2)):
    tmp = codon_table_v2[i].split()
    codon_table_dic[tmp[0]] = tmp[1]

print(codon_table_dic)
```
