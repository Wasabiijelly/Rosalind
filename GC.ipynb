```python
with open("C:\\Users\\ABCLAB\\Rosalind\\data\\rosalind_gc.txt",'r') as f:
    line = f.readlines()
    
print(line)
```

```python
id_idx = []

for i in range(len(line)):
    if line[i].find('>') != -1:
        id_idx.append(i)
        
print(id_idx)
print(len(id_idx))
```

```python
seq = []

for i in range(len(id_idx)-1):
    seq.append(''.join(line[id_idx[i]+1:id_idx[i+1]]))

seq.append(''.join(line[id_idx[len(id_idx)-1]+1:len(line)]))
print(len(seq))
print(seq)
print('-----------------------------------------------')
for i in range(len(seq)):
    seq[i] = seq[i].replace('\n',"")
print(seq)
```

```python
percent = []

for i in range(len(seq)):
    percent.append(((seq[i].count('C') + seq[i].count('G')) / len(seq[i])) * 100)
   
print(percent)
print(len(percent))
```

```python
max_idx = percent.index(max(percent))

print(max_idx)
```
